```python
import json
import matplotlib.pyplot as plt
import numpy as np

path = r"C:\Users\Linus\git\cryptCorr\MainProject\IMSniff_Testbed_OpenTAP\bin\Debug x64\Results\Recording_15_06_Evening\pcapAnalyzeResult_2023-06-26-17-55-14.json"

def min_max_normalize(data):
    min_val = min(data)
    max_val = max(data)
    normalized_data = [(x - min_val) / (max_val - min_val) for x in data]
    return normalized_data

# JSON-Daten laden
with open(path, 'r') as file:
    data = json.load(file)

# Listen für die Daten
total_sizes = []
total_counts = []
avg_sizes = []
capture_durations = []

# Daten aus dem JSON extrahieren
for entry in data:
    total_sizes.append(entry['totalSize'])
    total_counts.append(entry['totalCount'])
    avg_sizes.append(entry['avgSize'])
    capture_durations.append(entry['captureDuration'])

# Erstelle Subplots für die Boxplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

realDur = []
for dur in capture_durations:
    realDur.append(np.std(dur))

# Boxplot für totalSize
axes[0, 0].boxplot(total_sizes)
axes[0, 0].set_title('Total Size')
axes[0, 0].set_ylabel('Size [MB]')

# Boxplot für totalCount
axes[0, 1].boxplot(total_counts)
axes[0, 1].set_title('Total Count')

# Boxplot für avgSize
axes[1, 0].boxplot(avg_sizes)
axes[1, 0].set_title('Average Size')

# Boxplot für captureDuration
axes[1, 1].boxplot(capture_durations)
axes[1, 1].set_title('Capture Duration [Seconds]')

# Abstand zwischen den Subplots einstellen
plt.subplots_adjust(hspace=0.4)

# Grafik anzeigen
plt.show()
```

```python

# Listen für die normalisierten Größenwerte
normalized_sizes = []

# Maximale Größe ermitteln
max_size = 0
for file_name in file_names:
    with open(file_name, 'r') as file:
        data = json.load(file)
        size = data['totalSize']
        max_size = max(max_size, size)

# Größenwerte normalisieren
for file_name in file_names:
    with open(file_name, 'r') as file:
        data = json.load(file)
        size = data['totalSize']
        normalized_size = size / max_size
        normalized_sizes.append(normalized_size)

# Erstelle Boxplot der normalisierten Größenwerte
plt.boxplot(normalized_sizes)
plt.xlabel('Dateien')
plt.ylabel('Normalisierte Größe')
plt.title('Vergleich der Dateigrößen')
plt.xticks([1], ['Größe'])
plt.show()
```
