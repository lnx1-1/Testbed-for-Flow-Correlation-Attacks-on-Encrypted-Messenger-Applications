```python
import numpy as np
import matplotlib.pyplot as plt

labels = ['IMProxy Dataset', 'Erzeugte Testdaten']
values1 = [82.17, 100]
values2 = [0, 98.9]

# Erstelle Subplots
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# Linker Bargraph
ax[0].bar(labels, values1, width=0.4)  # Breite der Balken anpassen
ax[0].set_ylim(0, 110)  # Y-Achsenbereich anpassen
ax[0].set_ylabel('Prozent')
ax[0].set_title('Konsistente Dateigrößen')
# Werte über den Balken anzeigen
for i, v in enumerate(values1):
    ax[0].text(i, v, str(v) + '%', ha='center', va='bottom')

# Rechter Bargraph
ax[1].bar(labels, values2, width=0.35, align='center')  # Breite der Balken anpassen
ax[1].set_ylim(0, 110)  # Y-Achsenbereich anpassen
ax[1].set_ylabel('Prozent')
ax[1].set_title('Vorhandene lokale Dateien')
# Werte über den Balken anzeigen
for i, v in enumerate(values2):
    ax[1].text(i, v, str(v) + '%', ha='center', va='bottom')

# Abstand zwischen den Subplots einstellen
plt.subplots_adjust(wspace=0.3)

# Grafik anzeigen
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

labels = ['Konsistente Dateigrößen', 'Vorhandene lokale Dateien']
values1 = [82.17, 0]
values2 = [100, 98.9]

# Erstelle Subplot
fig, ax = plt.subplots(figsize=(8, 5))

# Position der Balken
bar_positions1 = np.arange(len(labels))
bar_positions2 = bar_positions1 + 0.4  # Ändere den Abstand zwischen den Balken hier

# Linker Bargraph
ax.bar(bar_positions1, values1, width=0.3, label='IMProxy Dataset')
# Werte über den Balken anzeigen
for i, v in enumerate(values1):
    ax.text(i, v, str(v) + '%', ha='center', va='bottom', fontsize=12)

# Rechter Bargraph
ax.bar(bar_positions2, values2, width=0.3, label='Testbed')
# Werte über den Balken anzeigen
for i, v in enumerate(values2):
    ax.text(i + 0.4, v, str(v) + '%', ha='center', va='bottom', fontsize=12)

# Achsenbeschriftung und Titel
ax.set_xticks(bar_positions1 + 0.2)
ax.set_xticklabels(labels)
ax.set_ylabel('Prozent')
# ax.set_title('Balkendiagramm')
ax.set_ylim(0, 110)

# Legende anzeigen
ax.legend(loc='upper left')

# Grafik anzeigen
plt.show()
```
